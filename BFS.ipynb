```python
def BFS(source_board, source_x, source_y,target_board):
    id_board = np.zeros((m,n))
    find_value = source_board[source_x][source_y]
    queue = []
    queue.append((source_x + 1, source_y))
    queue.append((source_x - 1, source_y))
    queue.append((source_x, source_y + 1))
    queue.append((source_x, source_y - 1))
    while len(queue) > 0:
        x, y = queue.pop(0)
        if x < 0 or x >= m or y < 0 or y >= n:
            continue
        if source_board[x][y] == find_value and source_board[x][y] != target_board[x][y]:
            return x,y
        id_board[x][y] = 1
        if id_board[x+1][y] == 0:
            queue.append((x + 1, y))
        if id_board[x-1][y] == 0:   
            queue.append((x - 1, y))
        if id_board[x][y+1] == 0:
            queue.append((x, y + 1))
        if id_board[x][y-1] == 0:
            queue.append((x, y - 1))
    return -1, -1
```
